```python
import requests
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
vaccination_path = "Vaccination Data/vaccination.csv"
vaccine_df = pd.read_csv(vaccination_path)
vaccine_df = vaccine_df.drop("Unnamed: 0", axis=1)
vaccine_df
```

```python
len(vaccine_df["Vaccine dose:1 "])
```

```python
x_axis = np.arange(0, len(vaccine_df["Vaccine dose:1 "]), 1)
plt.yscale('linear')

plt.ticklabel_format(style='plain')
plt.bar((x_axis - 0.4), vaccine_df["Vaccine dose:1 "], 0.2, color='red')
plt.bar((x_axis - 0.2), vaccine_df["Booster vaccine: 2"], 0.2, color='blue')
plt.bar(x_axis, vaccine_df["Booster vaccine: 3"], 0.2, color='green')
plt.bar((x_axis + 0.2), vaccine_df["Booster vaccine: 4"], 0.2, color='orange')
plt.xlabel("Provinces")
plt.ylabel("Canadians Vaccinated")
plt.legend(["1", "2", "3", "4"], title = "# of Vaccines")
plt.title("Total Vaccinations per Province")
plt.xticks(x_axis, ['AB', 'BC', 'MB', 'NB', 'NL', 'NS', 'NT', 'NU', 'ON', 'PE', 'QC', 'SK', 'YT'], fontsize=9)
plt.savefig("Output/VaccineTotals.png")
plt.show()
```

```python
plt.yscale('linear')
fig, ax = plt.subplots()
plt.ticklabel_format(style='plain')
plt.bar((x_axis - 0.4), vaccine_df["Vaccine coverage dose:1"], 0.2, color='red')
plt.bar((x_axis - 0.2), vaccine_df["Vaccine coverage booster:2"], 0.2, color='blue')
plt.bar(x_axis, vaccine_df["Vaccine coverage booster:3"], 0.2, color='green')
plt.bar((x_axis + 0.2), vaccine_df["Vaccine coverage booster:4"], 0.2, color='orange')
plt.xlabel("Provinces")
plt.ylabel("Percent of Canadians Vaccinated")
plt.legend(["1", "2", "3", "4"], title = "# of Vaccines")
plt.title("Vaccinated Percentages by Province")
ax.yaxis.set_major_formatter('{x:1.0f}%')
plt.xticks(x_axis, ['AB', 'BC', 'MB', 'NB', 'NL', 'NS', 'NT', 'NU', 'ON', 'PE', 'QC', 'SK', 'YT'], fontsize=9)
plt.savefig("Output/VaccinePercentage.png")
plt.show()
```
