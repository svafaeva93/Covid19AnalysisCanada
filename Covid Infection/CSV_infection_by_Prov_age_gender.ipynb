```python
import pandas as pd
```

```python
# Total cumulative cases (from 2020 to 2023) by province, Gender and Age Group

infection_df = pd.read_csv("Resources/Infection.dataset.csv")
infection_df.tail(10)

infection_df = infection_df[['GEO','Gender', 'Age group','Statistics','VALUE' ]]
infection_df = infection_df[infection_df['GEO']!= 'Canada']
infection_df = infection_df[infection_df['Gender']!= 'Total, all genders']
infection_df =infection_df[(infection_df['Age group']!= 'Total, all age groups')&
                                  (infection_df['Age group']!= 'Not stated, age group')]
infection_df =infection_df[(infection_df['Statistics']!= 'Total, by hospitalization status')&
                           (infection_df['Statistics']!= 'Total, by intensive care unit admittance')&
                           (infection_df['Statistics']!= 'Total, by death status')]
```

```python
infection_df_final= infection_df.groupby(['GEO','Gender', 'Age group']).sum()
```

```python
infection_df_final = infection_df_final.reset_index()
```

```python
infection_df_final
```

```python
# Infection cases by province

infection_df_province = infection_df_final[['GEO','VALUE']].groupby('GEO').sum().reset_index()
infection_df_province
```

```python
# Infection cases by Gender

infection_df_Gender = infection_df_final[['Gender', 'VALUE']].groupby('Gender').sum().reset_index()
infection_df_Gender
```

```python
# Infection cases by Age group

infection_df_Age = infection_df_final[['Age group', 'VALUE']].groupby('Age group').sum().reset_index()
infection_df_Age
```

```python
infection_df_final.to_csv('out.csv')
infection_df_final
```
