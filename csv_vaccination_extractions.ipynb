Dictionary:
    
cases_unvac_rate_per100K	The case rate per 100,000 for those unvaccinated. Calculated by dividing the number of cases among unvaccinated individuals by the total number of unvaccinated individuals. This rate is provided for the overall population as well as for each age group. 
cases_partial_vac_rate_per100K	The case rate per 100,000 for those partially vaccinated. Calculated by dividing the number of cases among partially vaccinated individuals by the total number of partially vaccinated individuals. This rate is provided for the overall population as well as for each age group. 
cases_notfull_vac_rate_per100K	The case rate per 100,000 for those not fully vaccinated. Calculated by dividing the number of cases among individuals not fully vaccinated by the total number of individuals not fully vaccinated. This rate is calculated for the overall population as well as for each age group.  
cases_full_vac_rate_per100K	The case rate per 100,000 for those fully vaccinated. Calculated by dividing the number of cases among fully vaccinated individuals by the total number of fully vaccinated individuals. This rate is provided for the overall population as well as for each age group. 
cases_boost_vac_rate_per100K	The case rate per 100,000 for those with at least one booster. Calculated by dividing the number of cases among individuals with at least one booster by the total number of individuals with at least one booster. This rate is provided for the overall population as well as for each age group. 
cases_unvac_rate_7ma	The average of the last seven days of the case rate per 100,000 for those unvaccinated
cases_partial_vac_rate_7ma	The average of the last seven days of the case rate per 100,000 for those partially vaccinated
cases_notfull_vac_rate_7ma	The average of the last seven days of the case rate per 100,000 for those not fully vaccinated
cases_full_vac_rate_7ma	The average of the last seven days of the case rate per 100,000 for those fully vaccinated
cases_boost_vac_rate_7ma	The average of the last seven days of the case reate per 100,000 for those with at least one booster

```python
vaccination_doses_df=pd.read_csv("Resources/vaccine_doses.csv")
vaccination_doses_df.tail(1) 


#the last rows basically gives us cumulative data for total_individuals_at_least_one	total_individuals_partially_vaccinated	total_doses_in_fully_vaccinated_individuals	total_individuals_fully_vaccinated	total_individuals_3doses
```

previous_day_total_doses_administered	Total number of doses administered during the previous day.
previous_day_at_least_one	Number of individuals who received their first dose during the previous day.
previous_day_fully_vaccinated	Number of individuals who became fully vaccinated the previous day.
previous_day_3doses	Number of individuals who became vaccinated with three doses the previous day.
total_doses_administered	Total number of doses administered to date.
total_individuals_at_least_one	Cumulative number of individuals who have received at least one dose.
total_individuals_partially_vaccinated	"Cumulative number of individuals who are partially vaccinated. Partially vaccinated is defined as receiving:
- the first of a two-dose vaccine series (e.g., Pfizer or Moderna), where the vaccine is Health Canada approved, or
- the first of a two-dose vaccine series, where the vaccine(s) is not Health Canada approved, or
- two doses of non-Health Canada approved vaccines, or
- one dose of a Health Canada-approved vaccine, followed by one dose of a non-Health Canada approved vaccine."
total_doses_in_fully_vaccinated_individuals	Total number of doses administered to individuals who had received two doses.  Discontinued as of December 3, 2021.
total_individuals_fully_vaccinated	"Cumulative number of individuals who are fully vaccinated. Fully vaccinated is defined as recieving: 
-one dose of Janssen (Johnson & Johnson), or
-two doses of any Health Canada approved vaccine, or
-one dose of a non-Health Canada approved vaccine, followed by one dose of a Health Canada-approved vaccine, or 
-three doses of any Health Canada approved vaccine, or 
-three doses of a vaccine, whether it is Health Canada approved or not."
total_individuals_3doses	"Cumulative number of individuals who are vaccinated with three doses. Vaccinated with three doses is defined as receiving three doses of any Health Canada approved vaccine. 
These individuals are a subset of the individuals captured in ""total_individuals_fully_vaccinated""."
